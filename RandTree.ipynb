```python
#Binary Tree Generation
def roots(L,L2):
    v = []
    for e in L:
        if L2[L.index(e)] == e + 1:
            v.append(e)
    return v

L = [1,2,3,4]
L2 = [2,1,4,3]

M = 4

for i in range(M):
    
    Rt = roots(L,L2)
    Rt.sort()
    while Rt != []:
        x = Rt[0]
        Rt.remove(x)
        for j in range(len(Rt)):
            Rt[j] += 4
        for k in range(len(L)):
            if L[k] >= x+1:
                L[k] += 4
        for k in range(len(L)):
            if L2[k] >= x+1:
                L2[k] += 4
        L.append(x+1)
        L.append(x+2)
        L2.append(x+2)
        L2.append(x+1) 
        L.append(x+3)
        L.append(x+4)
        L2.append(x+4)
        L2.append(x+3) 
        
        
zipped_lists = zip(L, L2)

sorted_pairs = sorted(zipped_lists)

tuples = zip(*sorted_pairs)

list1, list2 = [ list(tuple) for tuple in  tuples]
    
lines = []

for i in range(len(list1)):
    lines.append(str(list1[i])+" "+str(list2[i]))
    
f = open('RandTree8.txt', 'w')
for line in lines:
    f.write(line)
    f.write('\n')
```
