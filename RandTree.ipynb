```python
#Binary Tree Generation
def roots(L,L2):
    v = []
    for e in L:
        if L2[L.index(e)] == e + 1:
            v.append(e)
    return v

L = [1,2,3,4]
L2 = [2,1,4,3]

M = 4

for i in range(M):
    
    Rt = roots(L,L2)
    Rt.sort()
    while Rt != []:
        x = Rt[0]
        Rt.remove(x)
        for j in range(len(Rt)):
            Rt[j] += 4
        for k in range(len(L)):
            if L[k] >= x+1:
                L[k] += 4
        for k in range(len(L)):
            if L2[k] >= x+1:
                L2[k] += 4
        L.append(x+1)
        L.append(x+2)
        L2.append(x+2)
        L2.append(x+1) 
        L.append(x+3)
        L.append(x+4)
        L2.append(x+4)
        L2.append(x+3) 
        
        
zipped_lists = zip(L, L2)

sorted_pairs = sorted(zipped_lists)

tuples = zip(*sorted_pairs)

list1, list2 = [ list(tuple) for tuple in  tuples]
    
lines = []

for i in range(len(list1)):
    lines.append(str(list1[i])+" "+str(list2[i]))
    
f = open('RandTree8.txt', 'w')
for line in lines:
    f.write(line)
    f.write('\n')
```

```python
#Degree Three Vertex Tree Generation
def roots(L,L2):
    v = []
    for e in L:
        if L2[L.index(e)] == e + 1:
            v.append(e)
    return v

L = [1,2,3,4,5,6]
L2 = [2,1,4,3,6,5]


def Three_Vertex(Depths):
    for _ in range(len(Depths)):
        M = Depths[_]

        for i in range(M):
            Rt = roots(L,L2)
            Rt.sort()
            while Rt != []:
                x = Rt[0]
                Rt.remove(x)
                for j in range(len(Rt)):
                    Rt[j] += 4
                for k in range(len(L)):
                    if L[k] >= x+1:
                        L[k] += 4
                for k in range(len(L)):
                    if L2[k] >= x+1:
                        L2[k] += 4
                L.append(x+1)
                L.append(x+2)
                L2.append(x+2)
                L2.append(x+1) 
                L.append(x+3)
                L.append(x+4)
                L2.append(x+4)
                L2.append(x+3) 

        zipped_lists = zip(L, L2)

        sorted_pairs = sorted(zipped_lists)

        tuples = zip(*sorted_pairs)

        list1, list2 = [ list(tuple) for tuple in  tuples]

        lines = []

        for i in range(len(list1)):
            lines.append(str(list1[i])+" "+str(list2[i]))

        f = open("/Users/pantelistassopoulos/Downloads/ThreeVertexDepth"+str(M)+".txt", 'w')
        for line in lines:
            f.write(line)
            f.write('\n')
        f.close()
Three_Vertex([3])
```

```python
# add line segment left/ right

#Degree Three Vertex Tree Generation
def line(l, r):
    
    L = [1,2,3,4,5,6]
    L2 = [2,1,4,3,6,5]
    
    l_vertex = 3
    r_vertex = 5
    
    for _ in range(l):
        for α in range(len(L)):
            if L[α] > l_vertex:
                L[α] += 2
        for α in range(len(L2)):
            
            if L2[α] > l_vertex:
                L2[α] += 2
        L.append(l_vertex+1)
        L.append(l_vertex+2)
        L2.append(l_vertex+2)
        L2.append(l_vertex+1)
        l_vertex += 1
        r_vertex += 2
        
    
    for _ in range(r):
        for α in range(len(L)):
            if L[α] > r_vertex:
                L[α] += 2
        for α in range(len(L2)):
            
            if L2[α] > r_vertex:
                L2[α] += 2
        L.append(r_vertex+1)
        L.append(r_vertex+2)
        L2.append(r_vertex+2)
        L2.append(r_vertex+1)
        r_vertex += 1


    lines = []
    
    zipped_lists = zip(L, L2)

    sorted_pairs = sorted(zipped_lists)

    tuples = zip(*sorted_pairs)

    list1, list2 = [ list(tuple) for tuple in  tuples]
    for i in range(len(list1)):
        lines.append(str(list1[i])+" "+str(list2[i]))
        
    return lines
    

def lines(N):
    for n in range(N):
        f = open('/Users/pantelistassopoulos/Downloads/Line/Points_50/TreeLine50_'+str(n)+'_'+str(N-n)+'.txt', 'w')
    
        Lines = line(n, N-n)

        for x in Lines:
            f.write(x)
            f.write('\n')
        f.close()
    
lines(50)
```
